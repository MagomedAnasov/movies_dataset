# E2 - Insights about the director

## Movies is the main dataset, so it should be cleansed.

```python
# As movies is the main table, it should be cleansed and prepared for join. 
movies.drop(['Awards','DVD', 'Poster', 'Plot','Website', 'Production','Released','imdbVotes','Type','Rated', 'Runtime', 'Language'], axis = 1, inplace = True)
```

```python
movies.head()
```

```python
# Remove duplicates from movies
movies.drop_duplicates(subset='imdbID',inplace=True)
```

```python
movies.head()
```

```python
# Remove white spaces from imdbID column
movies['imdbID'] = movies['imdbID'].str.replace(' ', '')
```

```python
movies.head()
```

## As we do not want to include all movies, we should do some joins with rating and genres dataframes

### 1) Prepare rating dataset

```python
# Prepare rating dataframe
rating.head()
```

```python
# Remove duplicates
rating.drop_duplicates(subset='imdbID',inplace=True)
```

```python
rating.head()
```

```python
# Remove white spaces from imdbID column
rating['imdbID'] = rating['imdbID'].str.strip()
```

```python
rating.head()
```

### 2) Join movies and rating on imdbID

```python
# Left join on imdbID is needed 
rated_movies=pd.merge(movies,rating, how = 'left',on = 'imdbID')
```

```python
# Check data
rated_movies.head()
```

```python
# Remove movies with missing Rating in order to have only relevant movies.
rated_movies = rated_movies.dropna(subset=['Rating'])
```

```python
# Check if there are missing values in Director column
rated_movies.isna().sum(axis = 0)
```

```python
# We found out that there are 3.923 missing rows in Director column, so they need to be removed in order to have clean dataset
rated_movies = rated_movies.dropna(subset=['Director'])
```

```python
rated_movies.head()
```

```python
# Check information about new dataframe
rated_movies.info()
```

### 3) Prepare genres dataset

```python
# Check dataset
genres.head()
```

```python
# Remove duplicates
genres.drop_duplicates(subset='imdbID',inplace=True)
```

```python
genres.head()
```

```python
# Remove white spaces from imdbID 
genres['imdbID'] = genres['imdbID'].str.strip()
```

```python
# Remove white spaces from imdbID 
genres['Genre'] = genres['Genre'].str.strip()
```

```python
genres
```

### 4) Join rated_movies and genres on imdbID

```python
# Left join on imdbID is needed 
movies_final =pd.merge(rated_movies,genres, how = 'left',on = 'imdbID')
```

```python
movies_final.head()
```

```python
# Check if there are missing values in Genre column
movies_final.isna().sum(axis = 0)
```

```python
movies_final.head()
```

```python
# We found out that there are 2.289 NaNs in Genre column, so they need to be removed in order to have clean dataset
movies = movies_final.dropna(subset=['Genre'])
```

### In order to have only relevant genres we would exclude some of them

```python
# Take a look at distinct genres that are available in the dataset
genres['Genre'].unique()
```

#### Those would be Adult, Reality-TV, News, Talk-Show, Game-Show

```python
# Create a list with genres that need to be removed
black_list = ['Adult','Reality-TV','News','Talk-Show','Game-Show']
```

```python
# Remove rows where Genre is one of genres from black list
movies = movies_final[~movies_final.Genre.isin(black_list)]
```

```python
movies
```

```python
# Use groupby to identify directors with most produced movies
directors = movies.groupby('Director')['imdbID'].count().sort_values(ascending=False)
```

```python
directors[:10]
```

```python
directors=movies['Director'].value_counts()[:10]
pd.DataFrame({'Director':directors.index, 'NumberOfMoviesProduced':directors.values})
```
